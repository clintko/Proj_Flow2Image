# Set Environment

```python
### Basic
import numpy  as np
import scipy  as sc
import pandas as pd
import glob, os, re
import pickle, sys

### import flowImg
py_path = "/local_data/notebooks/clintko/GitRepo/Proj_Flow2Image"
sys.path.insert(0, py_path)
from flowImg import *

### global const
SEED    = 123
DAT_DIR = "/data/clintko/flow_EQAPOL_normal/mpsk"
```

# Load data

```python
!ls /data/clintko/flow_EQAPOL_normal/mpsk
```

```python
fname = "wrap01_mpsk.P"
fpath = os.path.join(DAT_DIR, fname)
file  = open(fpath, "rb")

wrap_mpsk = pickle.load(file)
file.close()
```

```python
print(wrap_mpsk)
```

# Dimensional reduction

```python
%%time
tsne = Transform_TSNE(n_components = 2, n_jobs = 14, random_state = SEED)
wrap_tsne = tsne.transform(wrap_mpsk, verbose=True)
```

```python
print(wrap_tsne)
```

# Store the results

```python
DAT_DIR = "/data/clintko/flow_EQAPOL_normal/mpsk"
```

```python
### set up a pickle file
fname = "wrap02_tsne_1e4_N10.P"
fpath = os.path.join(DAT_DIR, fname)
file = open(fpath, 'wb')

### store the dictionary
pickle.dump(wrap_tsne, file)
file.close()
```
