# Wrap the data

# Standardization

standardize the samples based on the first one

```python
scaler = Transform_StandardScaler(k = 0)
scaler = scaler.fit(wrap_data, verbose = True)
wrap_scaled = scaler.transform(wrap_data)
```

helper function to check the mean and std before and after

```python
def get_mean_std_samples(dat):
    """get the mean and standard devision"""
    samples = [dat.get_sample(idx) for idx in range(dat.get_num_sample())]
    mu = np.array([sample.mean(axis = 0) for sample in samples])
    sd = np.array([sample.std(axis = 0)  for sample in samples])
    return (mu, sd)
```

```python
mu, sd = get_mean_std_samples(wrap_data)
print("Mean")
print(str(np.round(mu, decimals=2)))
print("=======================")
print("SD")
print(str(np.round(sd, decimals=2)))
```

```python
mu, sd = get_mean_std_samples(wrap_scaled)
print("Mean")
print(str(np.round(mu, decimals=2)))
print("=======================")
print("SD")
print(str(np.round(sd, decimals=2)))
```
