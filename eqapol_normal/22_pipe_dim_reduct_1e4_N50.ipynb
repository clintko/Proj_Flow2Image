# Set Environment

```python
### Basic
import numpy  as np
import scipy  as sc
import pandas as pd
import glob, os, re
import pickle, sys

### import flowImg
py_path = "/local_data/notebooks/clintko/GitRepo/Proj_Flow2Image"
sys.path.insert(0, py_path)
from flowImg import *

### global const
SEED    = 123
DAT_DIR = "/data/clintko/flow_EQAPOL_normal/pipe01"
```

# Load data

```python
!ls /data/clintko/flow_EQAPOL_normal/pipe01
```

```python
fname = "wrap03_subs_1e4_N50.P"
fpath = os.path.join(DAT_DIR, fname)
file  = open(fpath, "rb")

wrap_subs = pickle.load(file)
file.close()
```

```python
print(wrap_subs)
```

# Dimensional reduction

```python
%%time
tsne = Transform_TSNE(n_components = 2, n_jobs = 14, random_state = SEED)
wrap_tsne = tsne.transform(wrap_subs, verbose=True)
```

```python
print(wrap_tsne)
```

# Store the results

```python
### set up a pickle file
fname = "wrap04_tsne_1e4_N50.P"
fpath = os.path.join(DAT_DIR, fname)
file = open(fpath, 'wb')

### store the dictionary
pickle.dump(wrap_tsne, file)
file.close()
```
