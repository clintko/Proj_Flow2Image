```python
def generate_sample(mean, cov, num_events):
    """generate a guassian mixture sample"""
    ### check input
    assert mean.shape[0] == cov.shape[0]
    assert mean.shape[0] == num_events.shape[0]
    
    ### number of clusters
    num_clusters = mean.shape[0]
    
    ### generate multiple clusters; each cluster has #num_events points
    dat = [np.random.multivariate_normal(
               mean[idx], cov[idx], num_events[idx]) 
           for idx in range(num_clusters)]
    
    ### combine the results
    dat = np.concatenate(dat)
    return dat
```
